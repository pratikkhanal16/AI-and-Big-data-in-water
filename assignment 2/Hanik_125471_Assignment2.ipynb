# Hydrological cycle 

The hydrological cycle, also known as the water cycle, is a complex and continuous process that describes the movement and transformation of water within the Earth's atmosphere, land surfaces, and oceans (Oki and Kanae, 2006). This cycle is driven by solar energy and gravity, and it plays a crucial role in shaping our planet's climate, ecosystems, and water resources (Trenberth, Fasullo and Kiehl, 2009

## Components of the hydrological cycle

**1. Precipitation:**
This is the main input of water into the terrestrial part of the cycle. Precipitation occurs in various forms such as rain, snow, sleet, or hail. When precipitation falls, it follows several pathways:

**2. Interception:**
Also known as canopy water storage, this occurs when vegetation, particularly forests, intercepts rainfall before it reaches the ground. Some of this water evaporates directly back into the atmosphere, while the rest eventually reaches the soil through throughfall or stemflow (Crockford and Richardson.

**3. Infiltration and Soil Moisture Storage:**
A portion of the precipitation that reaches the ground infiltrates into the soil. This water is stored in the unsaturated zone of the soil, where it's available for plant uptake and evaporation.

**4. Surface Runoff:**
When precipitation exceeds the infiltration capacity of the soil, excess water flows over the land surface into streams, rivers, and eventually to the oceans. This process is influenced by factors such as rainfall intensity, soil properties, and land cover.

**5. Groundwater Recharge and Storage:**
Water that percolates deeply into the soil can reach the saturated zone, replenishing groundwater aquifers. This stored water can resurface through springs or be extracted through wells.

**6. Subsurface Flow:**
Water in the soil can move laterally, contributing to streamflow. This process, known as interflow or subsurface runoff, is particularly important in maintaining baseflow in streams during dry periods.

**7. Evaporation:**
This process occurs from various water bodies (oceans, lakes, rivers) and land surfaces. Solar energy causes water to change from liquid to vapor, returning to the atmosphere.

**8. Transpiration:**
Plants absorb water from the soil and release it as vapor through their leaves. This process is crucial for plant growth and contributes significantly to the water cycle, especially in forested areas.

**9. Evapotranspiration:**
This term combines evaporation and transpiration, representing the total water vapor flux from land to atmosphere.

**10. Atmospheric Transport:**
Water vapor in the atmosphere is transported by winds, redistributing moisture across the globe. This process is key to understanding why some regions receive more precipitation than others (Gimeno et al., 2012).

**11. Condensation:**
As water vapor rises and cools in the atmosphere, it condenses to form clouds, setting the stage for precipitation and completing the cycle (Lamb and Verlinde, 2011).

It's important to note that the hydrological cycle operates at various spatial and temporal scales. While some processes like precipitation and evaporation can occur rapidly, others like groundwater movement can take years or even centuries (Gleeson et al., 2016). Understanding the hydrological cycle is crucial for managing water resources, predicting floods and droughts, and assessing the impacts of climate change on water availability. The cycle is intricately linked with energy exchanges in the Earth system, influencing and being influenced by climate patterns (IPCC, 2021).

```python
'''This section is for importing all the necessary libraries'''
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
'''This cell creates a function that can be used to load different component (parameters) text files of the hydrological cycle
and calculate statistical values such as mean, median, standard deviation, 25th and 75th percentile'''

def stat_basin(parameters, unit):
    data = np.loadtxt(parameters,skiprows =1) # Load txt files
    year = data[:,0] 
    month = data[:,1]
    chao = data[:,2]
    mun = data[:,3]
    
    mean_chao = np.mean(chao)
    print("The mean " + parameters[:-4]+ " of the Chaophraya river basin is", mean_chao, unit)
    mean_mun = np.mean(mun)
    print("The mean " + parameters[:-4]+ " of the Mun river basin is", mean_mun, unit)
    
    std_chao = np.std(chao)
    print("The standard deviation of " + str(parameters[:-4])+ " of the Chaophraya river basin is", std_chao, unit)
    std_mun = np.std(mun)
    print("The standard deviation of " + parameters[:-4]+ " of the Mun river basin is", std_mun, unit)

    median_chao = np.median(chao)
    print("The median of " + parameters[:-4]+ " of the Chaophraya river basin is", median_chao, unit)
    median_mun = np.median(mun)
    print("The median of " + parameters[:-4]+ " of the Mun river basin is", median_mun, unit)
    
    percentile_25_chao = np.percentile(chao, 25)
    print("The 25th percentile " + parameters[:-4]+ " of the Chaophraya river basin is", percentile_25_chao, unit)
    percentile_25_mun = np.percentile(mun, 25)
    print("The 25th percentile " + parameters[:-4]+ " of the Mun river basin is", percentile_25_mun, unit)

    percentile_75_chao = np.percentile(chao, 75)
    print("The 75th percentile " + parameters[:-4]+ " of the Chaophraya river basin is", percentile_75_chao, unit)
    percentile_75_mun = np.percentile(mun, 75)
    print("The 75th percentile " + parameters[:-4]+ " of the Mun river basin is", percentile_75_mun, unit)
    
    return mean_chao, mean_mun, std_chao, std_mun, median_chao, median_mun, percentile_25_chao, percentile_25_mun, percentile_75_chao, percentile_75_mun
```

```python
'''This section creates a empty dataframe to store all the statistical values of different parameters for both the basins'''
stat_df = pd.DataFrame()
stat_df['basin_name'] = ['chao', 'mun', 'chao', 'mun', 'chao', 'mun', 'chao', 'mun', 'chao', 'mun']
stat_df['statistics'] = ['mean', 'mean', 'std', 'std', 'median', 'median', 'percentile_25', 'percentile_25', 'percentile_75', 'percentile_75']
print(stat_df)
```

```python
'''This section calculates the statistics for evapotranspiration for both basins and plots a bar graph for comparison'''
mean_chao, mean_mun, std_chao, std_mun, median_chao, median_mun, percentile_25_chao, percentile_25_mun, percentile_75_chao, percentile_75_mun = stat_basin('evapotranspiration.txt','mm/s')
ET_stat = [mean_chao, mean_mun, std_chao, std_mun, median_chao, median_mun, percentile_25_chao, percentile_25_mun, percentile_75_chao, percentile_75_mun]
stat_df['ET_stat'] = ET_stat

plt.figure(figsize = (10,6), dpi=500)
sns.barplot(x = stat_df['statistics'], y = stat_df['ET_stat'], hue = stat_df['basin_name'])
plt.xlabel('Statistics', fontsize = 14)
plt.ylabel('Evapotranspiration (mm/s)', fontsize = 14)
plt.title('Statistical comparision of Evapotranspiration of Chao Phraya and Mun river basin', fontsize = 18)
plt.legend(loc= 'upper left')
plt.tight_layout()
```

### *Statistical comparison of evapotranspiration*

The **Chaophraya River Basin** has a **higher** mean, 25th, and 75th percentile evapotranspiration values, indicating higher evapotranspiration overall, but with **low standard deviation** indicating less variability compared to the Mun River Basin. However, the median value of **Mun river basin** is **higher** than that of **Chaophraya River Basin**.

```python
'''This section calculates the statistics for plant transpiration for both basins and plots a bar graph for comparison'''
mean_chao, mean_mun, std_chao, std_mun, median_chao, median_mun, percentile_25_chao, percentile_25_mun, percentile_75_chao, percentile_75_mun = stat_basin('plant_traspiration.txt','mm/s')
PT_stat = [mean_chao, mean_mun, std_chao, std_mun, median_chao, median_mun, percentile_25_chao, percentile_25_mun, percentile_75_chao, percentile_75_mun]
stat_df['PT_stat'] = PT_stat
print(stat_df)

plt.figure(figsize = (10,6), dpi=500)
sns.barplot(x = stat_df['statistics'], y = stat_df['PT_stat'], hue = stat_df['basin_name'])
plt.xlabel('Statistics', fontsize = 14)
plt.ylabel('Plant transpiration (mm/s)', fontsize = 14)
plt.title('Statistical comparision of Plant transpiration of Chao Phraya and Mun river basin', fontsize = 18)
plt.legend(loc= 'upper left')
plt.tight_layout()
```
